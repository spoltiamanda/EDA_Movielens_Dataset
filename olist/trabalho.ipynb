Importando o arquivo com as infos do pedido CDD - Closed Deals Dataset

Explorando os arquivos

# #  criar tabela com todos os MQLs, com todas as informações disponíveis para cada um

Juntando os dois arquivos num só, através de um left join. Pois a necessidade aqui é ter todos os registros da base mql adicionando os da base cdd quando houver match

```python
df=pd.merge(mql, cdd, on="mql_id", how="left")
df=df.drop_duplicates() #tirando possíveis linhas duplicadas
df.head(3)
```

```python
# verificando se o arquivo tem pelo menos a mesma quantidade de registros que o mql
df['mql_id'].count()
```

```python
#verificando se o join "deu certo" ou seja, se existe pelo menos uma informação do cdd em mql
df[df['seller_id'].notnull()].head(5)
```

# #criar uma coluna com quais MQLs fecharam acordo e quais foram perdidos.

Para esta etapa, vou considerar que o acordo foi fechado quando won_date é diferente de nulo

```python
df['status_acordo'] = np.where(df['won_date'].notnull(), 'fechado', 'perdido')
```

```python
df[df['won_date'].notnull()].head(5) 
```

## Perguntas a serem respondidas

Perguntas a serem respondidas:
- Qual foi a taxa de conversão total?
- Qual foi a taxa de conversão de cada origem?
- Qual foi a taxa de conversão de cada página inicial?
- Para cada SR, qual a receita média declarada dos leads?
- Para cada SR, quantos de cada business_type a pessoa converteu?
- Para cada SR, quantos de cada lead_behaviour_profile a pessoa converteu?
- Para cada SR, quantos de cada lead_type a pessoa converteu?

```python
#Qual foi a taxa de conversão total? 
total_convertidos=df[df['status_acordo']=='fechado'].count()['mql_id']
total_leads=df['mql_id'].count()
print('taxa de conversão =',total_convertidos/total_leads*100,'%')
```

```python
df.head()
```

```python
#Qual foi a taxa de conversão de cada origem?
```

```python
taxa_por_origem=df.groupby('origin')['status_acordo'].count().to_frame('total')
taxa_por_origem['fechado']=df[df['status_acordo']=='fechado'].groupby('origin')['status_acordo'].count()
taxa_por_origem['tx_%']=taxa_por_origem['fechado']/taxa_por_origem['total']*100
```

```python
taxa_por_origem
```

```python
#Qual foi a taxa de conversão de cada página inicial?
taxa_por_LP=df.groupby('landing_page_id')['status_acordo'].count().to_frame('total')
taxa_por_LP['fechado']=df[df['status_acordo']=='fechado'].groupby('landing_page_id')['status_acordo'].count()
taxa_por_LP['tx_%']=taxa_por_LP['fechado']/taxa_por_LP['total']*100
```

```python
taxa_por_LP
```

```python
#Para cada SR, qual a receita média declarada dos leads? sr_id
df.groupby('sr_id')['declared_monthly_revenue'].mean().to_frame('Receita_media_declarada')
```

```python
#Para cada SR, quantos de cada business_type a pessoa converteu?
df[df['status_acordo']=='fechado'].groupby(['sr_id','business_type'])['business_type'].count().to_frame('Qtd')
```

```python
#Para cada SR, quantos de cada lead_behaviour_profile a pessoa converteu?
df[df['status_acordo']=='fechado'].groupby(['sr_id','lead_behaviour_profile'])['lead_behaviour_profile'].count().to_frame('Qtd')
```

```python
#Para cada SR, quantos de cada lead_type a pessoa converteu?
df[df['status_acordo']=='fechado'].groupby(['sr_id','lead_type'])['lead_type'].count().to_frame('Qtd')
```
